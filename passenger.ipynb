Problem Statement

Air travel demand changes over time due to seasonality, economic activity, holidays, tourism, and business activities. Airlines need to understand passenger behavior to improve scheduling, forecasting, pricing, staffing, and resource allocation.

This project analyzes an air passenger dataset to identify trends, seasonal patterns, and important statistical insights that can support business decision-making.

## Objectives

The main objectives of this project are to:

1. Analyze historical air passenger traffic data.
2. Identify trends and seasonal patterns in passenger movement.
3. Compute descriptive statistical measures such as mean, median, and percentiles.
4. Visualize passenger traffic using charts and plots.
5. Detect unusual passenger patterns and possible outliers.
6. Analyze yearly and monthly passenger behavior.
7. Use moving averages to identify long-term trends.
8. Generate business insights from the dataset.
9. Provide recommendations for airline planning and decision-making.
10. Understand how data analysis supports forecasting and operational management in the airline industry.

```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset

df = pd.read_csv('air_passengers.csv')
```

```python
#Data cleaning

df.columns = ['Month', 'Passengers']

df['Month'] = pd.to_datetime(df['Month'], format='%Y-%m')

df = df.sort_values('Month')
```

```python
#Data inspection

print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
#Trend visualization

fig, ax = plt.subplots(figsize=(12,5))

ax.plot(df['Month'], df['Passengers'])

ax.set_title('Air Passenger Trend (1949–1960)')
ax.set_xlabel('Year')
ax.set_ylabel('Passengers')

ax.grid(True)

plt.show()
```

```python
#Yearly analysis

df['Year'] = df['Month'].dt.year

yearly = df.groupby('Year')['Passengers'].sum()
print(yearly)

#yearly trend
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(yearly.index, yearly.values, marker='o')

ax.set_title('Yearly Passenger Trend')
ax.set_xlabel('Year')
ax.set_ylabel('Total Passengers')

ax.grid(True)

plt.show()
```

```python
#Monthly patterns

df['Month_Name'] = df['Month'].dt.month_name()

monthly_avg = df.groupby('Month_Name')['Passengers'].mean().sort_values()

print(monthly_avg)
```

```python
#Descriptive Statistics

print("Mean:", df['Passengers'].mean())
print("Median:", df['Passengers'].median())
print("Std Dev:", df['Passengers'].std())
```

```python
#90th percentile

p90 = np.percentile(df['Passengers'], 90)
print("90th Percentile:", p90)
```

```python
#Distribution(Histogram)

fig, ax = plt.subplots()

ax.hist(df['Passengers'])

ax.set_title('Distribution of Passenger Counts')
ax.set_xlabel('Passengers')
ax.set_ylabel('Frequency')

plt.show()
```

```python
#Boxplot

fig, ax = plt.subplots()

ax.boxplot(df['Passengers'])

ax.set_title('Passenger Outliers Detection')

plt.show()
```

```python
#Moving averages

df['Moving_Avg'] = df['Passengers'].rolling(window=12).mean()

fig, ax = plt.subplots(figsize=(12,5))

ax.plot(df['Month'], df['Passengers'], label='Original')
ax.plot(df['Month'], df['Moving_Avg'], label='12-Month Moving Average')

ax.set_title('Passenger Trend with Moving Average')
ax.set_xlabel('Year')
ax.set_ylabel('Passengers')

ax.legend()

plt.show()
```

## Key Insights

1. Air passenger traffic shows a strong upward trend from 1949 to 1960.

3. Clear seasonal patterns exist with peaks during certain months every year.
4. Variability increases over time, meaning demand is becoming more volatile.
5. The 90th percentile shows periods of very high passenger demand.
6. Moving average confirms long-term steady growth without major decline.
